```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fig.mplstyle')
save_dir = "/home/supercgor/gitfile/texfile/Modern_Physics_Lab/7.1/fig"
data_name = "data.xlsx"
from scipy.optimize import curve_fit
```

```python
A = pd.read_excel(f"{save_dir}/{data_name}", 0)
B = pd.read_excel(f"{save_dir}/{data_name}", 1)
C = pd.read_excel(f"{save_dir}/{data_name}", 2)
Aa = A.iloc[:,0].to_numpy()
Ab = A.iloc[:,1].to_numpy()
Ac = A.iloc[:,2].to_numpy()
Ad = A.iloc[:,3].to_numpy()
Ad = Ad/np.nanmax(Ad)
Ba = B.iloc[:,0].to_numpy()
Bb = B.iloc[:,0].to_numpy()
Bb = Bb/np.nanmax(Bb)
Ca = C.iloc[:,0].to_numpy()
Cb = C.iloc[:,1].to_numpy()
Ca = Ca/np.nanmax(Ca)
Cc = Cb - np.nanmin(Cb)
Cc = np.sin(2*np.pi*Cc/49.29)
Cc = np.clip(Cc, 0, np.inf)
```

```python
a = plt.subplot(131)
b = plt.subplot(132)
c = plt.subplot(133)
# 拟合曲线
Aaa = np.linspace(Aa.min(),Aa.max(), 100)#插入点
# 插值函数
# fa = interp1d(Aa, Ab, kind="cubic")
# fb = interp1d(Aa, Ac, kind="cubic")
# fc = interp1d(Aa, Ad, kind="cubic")
# 画结果
a.plot(Aa, Ab, color="blue",linewidth=1, marker="s", markerfacecolor="white", markersize=4, label="工作电流")
b.plot(Aa, Ac, color="red",linewidth=1, marker="s", markerfacecolor="white", markersize=4, label="频率")     
c.plot(Aa, Ad, color="green",linewidth=1, marker="s", markerfacecolor="white", markersize=4, label="相对功率")
# 名称
a.set_xlabel("$U$ (V)")
b.set_xlabel("$U$ (V)")
c.set_xlabel("$U$ (V)")
a.set_ylabel("$I$ (mA)")
b.set_ylabel("$f$ (GHz)")
c.set_ylabel("$P$")
# title
a.set_title("(a)", x=-0.1)
b.set_title("(b)", x=-0.1)
c.set_title("(c)", x=-0.1)
# 显示label
a.legend()
b.legend()
c.legend()
plt.tight_layout()
```

```python
plt.rcParams["figure.figsize"] = (5,3.75)
plt.plot(Ba,Bb, color="blue",linewidth=1, marker="s", markerfacecolor="white", markersize=4, label="相对功率")
plt.xlabel("$f$ (GHz)")
plt.ylabel("$P$")
plt.legend()
plt.tight_layout()
```

```python
### define the fit functions, y = a * x^b ###
def target_func(x, a, b):
    return a*(x**b)
 
### curve fit ###
popt, pcov  = curve_fit(target_func, Cc, Ca)
 
### Calculate R Square ###
calc_ydata = [target_func(i, popt[0], popt[1]) for i in Cc]
res_ydata  = np.array(Ca) - np.array(calc_ydata)
ss_res     = np.sum(res_ydata**2)
ss_tot     = np.sum((Ca - np.mean(Ca))**2)
r_squared  = 1 - (ss_res / ss_tot)
print(f"a = {popt[0]}  b = {popt[1]}   R2 = {r_squared}")
```
